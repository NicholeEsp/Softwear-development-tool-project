```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars= pd.read_csv('../vehicles_us.csv')
```

```python
display(cars)
```

```python
display(cars.price.info())
```

```python
display(cars.days_listed.info())
display(cars.days_listed.value_counts())
display(cars.days_listed.value_counts().sort_index())
display(cars.days_listed.value_counts().sort_index().plot( xlabel='Number of days on Market', ylabel='Number of vehicles',
                                                          title='How many days a vehicle had been listed for'))
```

```python
import numpy as np
```

```python
condition=cars.condition
```

```python
display(condition.isnull().sum())
```

```python
display(condition.unique())
days_listed_count = cars.value_counts(subset=["condition", "days_listed"]).to_frame().reset_index()
days_listed_count.pivot(columns=["condition"], index=["days_listed"], values="count").plot.line(title='Length on market vs quality of car',
                        xlabel='Days Listed', ylabel='amount of cars')

```
